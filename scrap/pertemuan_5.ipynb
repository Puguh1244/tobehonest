```python
import pandas as pd
import nltk
from nltk import *
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
import re
import string
from collections import Counter

df= pd.read_csv('data_teks.csv')
df.head()
```

```python
df['Tweet']=df['Tweet'].str.lower()
df
```

```python
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

df['Tweet'] = df['Tweet'].apply(lambda tweet: remove_punctuation(tweet))
df.head()
```

```python
def remove_url(text):
    url=re.compile(r'https?://\S+|www.\S+')
    return url.sub(r'',text)
df['Tweet'] = df['Tweet'].apply(lambda tweet: remove_url(tweet))
df
```

```python
nltk.download('stopwords')
stop_indo=set(stopwords.words('indonesian'))

def remove_stopwords (text):
    filter_word=[word for word in str(text).split() if word.lower() not in stop_indo]
    return " ".join(filter_word)
df['Tweet'] = df['Tweet'].apply(lambda tweet: remove_stopwords(tweet))
df
```

```python

cnt= Counter()
for text in df['Tweet'].values:
    for word in text.split():
        cnt[word]+=1
cnt.most_common() 
```

```python
penghapusan= set([w for (w, wc) in cnt.most_common(10)]) # wa adalah kata dan wc adalah banyak kemunculan kata dalam dataset

def remove_penghapusan(text):
    return " ".join([word for word in str(text).split()if word not in penghapusan])
df['penghapusan']= df['Tweet'].apply(lambda text: remove_penghapusan(text))
df
```

```python
rare_words=10
rare= set([w for (w,wc) in cnt.most_common()[:-rare_words-1:-1]])

def remove_rare(text):
    return " ".join([word for word in str(text).split() if word not in rare])
df['rare']=df['Tweet'].apply(lambda text : remove_rare(text))
df.drop('Date',axis=1)
```

```python
import emoji

def remove_emoji(text):
    return emoji.replace_emoji(text," ")
df['no_emoji']=df['Tweet'].apply(lambda text : remove_emoji(text))
df
```

```python
def remove_emoticon(text):
    emoticon_pattern = re.compile(r'[:;=8][\-~]?[)DdpP\(\[\]/\\]')
    return emoticon_pattern.sub(r'', text)
df['no_emoticon']=df['Tweet'].apply(lambda text:remove_emoticon(text))
df
```

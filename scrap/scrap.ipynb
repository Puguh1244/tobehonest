```python
contents = page.content
print(contents)
```

```python
cont = contents.decode('utf-8') # untuk mengubah huruf menjadi utf 8
cont
```

```python
print(cont[:5000])  # Cetak 500 karakter pertama untuk memastikan karakter yang dipakai oleh html
```

```python
import re
import pandas as pd 

# Mengambil data dari tag <p>
titles = re.findall(r'<p>(.*?)</p>', cont)

# Pembenahan tanda baca
titles = [title.replace('&#039;', "'") for title in titles]
titles = [title.replace('&#x101;', "") for title in titles]
titles = [title.replace('i&#x17C;', "") for title in titles]
titles = [title.replace('&#x1E63;', "") for title in titles]
titles = [title.replace('&#x12B;', "") for title in titles]
titles = [title.replace('&#x1E6D;', "") for title in titles]

# Filter berdasarkan angka, tidak ada '//' dan '<'
filtered_titles = [title for title in titles if re.search(r'\d+', title) and '//' not in title and '<' not in title]

# Memformat angka terakhir yang bergandengan dengan huruf menjadi subscript
formatted_titles = [re.sub(r'(\d+)(\D*$)', r'<sub>\1</sub>\2\n', title) for title in filtered_titles]

# Menampilkan hasil filtered titles
print(formatted_titles)

# Mengonversi hasil yang diformat ke DataFrame
df = pd.DataFrame(formatted_titles, columns=['bacaan']) 

df.to_csv('yaasiin.csv', index=False, encoding='utf-8')

print("Data telah diformat dan disimpan ke 'yaasiin.csv'")
df
```

```python
# splitting 
korban=data.iloc[:,[0]]
faktor_pengemudi=data.iloc[:,[1]]
faktor_kendaraan=data.iloc[:,[3]]
faktor_jalan=data.iloc[:,[2]]
```

```python
X3 = data[['Faktor Pengemudi (orang)', 'Faktor Jalan (kasus)', 'Faktor Kendaraan (kasus)']]
y3 = data[['Jumlah Kecelakaan Lalu Lintas (kasus)']]
```

```python
regressor_pengemudi=LinearRegression()
regressor_jalan=LinearRegression()
regressor_kendaraan=LinearRegression()
regressor_mean=LinearRegression()
regressor_pengemudi.fit(korban,faktor_pengemudi)
regressor_kendaraan.fit(korban,faktor_kendaraan)
regressor_jalan.fit(korban,faktor_jalan)
regressor_mean.fit(X3,y3)
```

```python
# Misalkan kita memiliki koefisien regresi untuk berbagai faktor
coef_jalan = regressor_jalan.coef_
coef_pengemudi = regressor_pengemudi.coef_
coef_kendaraan = regressor_kendaraan.coef_

intecep_jalan=regressor_jalan.intercept_
intercep_kendaraan=regressor_kendaraan.intercept_
intercep_pengemudi=regressor_pengemudi.intercept_

rjalan=regressor_jalan.score(korban,faktor_jalan)
rkendaraan=regressor_kendaraan.score(korban,faktor_kendaraan)
rpengemudi=regressor_pengemudi.score(korban,faktor_pengemudi)

# Menggunakan f-string untuk menampilkan hasil dalam satu baris
print(f'Koefisien Faktor Jalan:     {coef_jalan} | '
      f'Intercept Faktor Jalan: {intecep_jalan} |'
      f'R^ Jalan : {intercep_pengemudi}')
print(f'Koefisien Faktor Pengemudi: {coef_pengemudi} | '
      f'Intercept Faktor Pengemudi: {intercep_pengemudi}')
print(f'Koefisien Faktor Kendaraan: {coef_kendaraan} | '
      f'Intercept Faktor Kendaraan: {intercep_kendaraan}')

print()
print()
print(f'R^ jalan: {rjalan}  |'
      f'R^ Pengemudi: {rpengemudi} |'
      f'R^ Kendaraan: {rkendaraan}')

```
